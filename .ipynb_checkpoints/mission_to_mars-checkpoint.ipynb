```python
# import dependencies

import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from splinter import Browser
```

## Scrape the NASA Mars News Site and collect the latest News Title and Paragraph Text

```python
# scrape mars.nasa.gov/news into a beautiful soup object
url = 'http://mars.nasa.gov/news'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#print the soup object to review it and make sure we are good to go.

print(soup.prettify())
```

```python
# Collect the latest news title from the soup object

news_title = soup.find_all('div', class_='content_title')[0].a.text.strip()
```

```python
# Collect the news paragraph blurb from the soup object

news_paragraph = soup.find_all('div', class_='rollover_description_inner')[0].text.strip()
```

## Scrape the JPL Featured Space Image site and find the image URL for the current featured Mars image

```python
# scrape https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars into a beautful soup object

url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Print the soup object to review the HTML

print(soup.prettify)
```

```python
# Fetch the URL for the image
baseurl = 'https://www.jpl.nasa.gov/spaceimages'

# I feel like there is an easier way to do this part, but it worked. Oh, i was supposed to use splinter.
relativeurl = soup.find_all('article', class_='carousel_item')[0]['style'].split(' ')[1].split('(')[1].replace("'", '').replace(")", '').replace(";", '')

featured_image_url = baseurl + relativeurl
```
